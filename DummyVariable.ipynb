```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
data=[
("newjersey",2600,550000),
("newjersey",3200,610000),
("newjersey",3600,680000),
("newjersey",4000,725000),
("california",2600,850000),
("california",2800,900000),
("california",3300,925000),
("california",3600,985000),
("georgia",2600,350000),
("georgia",2900,385000),
("georgia",3100,405000),
("georgia",3600,430000)]
df=pd.DataFrame(data,columns=["town","area","price"])
df      
```

```python
dummy=pd.get_dummies(df.town,dtype=int)
```

```python
dummy
```

```python
merged=pd.concat([df,dummy],axis='columns')
merged
```

```python
final=merged.drop(['town','georgia'],axis='columns')
final
```

```python

model=linear_model.LinearRegression()
```

```python
x=final.drop(['price'],axis='columns')
```

```python
y=final.price 
```

```python
model.fit(x,y)
```

```python
model.predict([[4000,0,1]])
```

```python
model.score(x,y)
```
